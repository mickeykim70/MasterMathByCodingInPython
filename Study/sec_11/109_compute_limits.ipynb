---
$$ \lim_{x \to a} f(x) $$
$$ \lim_{x \to 4} \frac{x^2}{2} $$

```python
import sympy as sym
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display, Math
```

```python
x = sym.symbols('x')

fx = x**3

lim_pnt = 1.5
lim = sym.limit(fx, x, lim_pnt)

# latex_str = f'\\lim_{{x \\to {lim_pnt}}} {sym.latex(fx)} = {lim}'
# latex_str

# Math( '\\lim_{x \\to %g} %s = %g' %(lim_pnt,sym.latex(fx),lim))
Math(f'\\lim_{{x \\to {lim_pnt}}} {sym.latex(fx)} = {lim}')
```

```python
# 극한값 계산
x = sym.symbols('x')
fx = x**3
lim_pnt = 1.5
lim = sym.limit(fx, x, lim_pnt)

# display 함수로 출력
display(sym.limit(fx, x, lim_pnt))

latex_expr = f'\\lim_{{x \\to {lim_pnt}}} {sym.latex(fx)} = {lim}'
display(Math(latex_expr))
```

```python
fxx = sym.lambdify(x, fx)
xx = np.linspace(-5, 5, 200)

plt.plot(xx,fxx(xx))
plt.plot(lim_pnt, lim, 'ro')

plt.show()
```

```python
fx = (x**2)/(x-2)
fxx = sym.lambdify(x, fx)
xx = np.linspace(1, 3, 102)

lim_pnt = 2
lim = sym.limit(fx, x, lim_pnt, dir='-')

plt.plot(xx,fxx(xx))
plt.plot(lim_pnt, lim, 'ro')

plt.show()

latex_expr = f'\\lim_{{x \\to {lim_pnt}^-}} {sym.latex(fx)} = {lim}'
display(Math(latex_expr))
```
