### Plotting Exercise

```python
import sympy as sym

x = sym.Symbol('x')
y = x**2 - 3*x

for point_i in range(-11, 12):
    plt.plot(point_i, y.subs(x, point_i), 'o')
    
plt.xlabel('x')
plt.ylabel(f'f(x) = {y}')
plt.show()
```

```python
x = np.arange(-20, 21, 1)

for xi in x:
    plt.plot([0, xi],[0, np.abs(xi)**(1/2)])
    
plt.show()
```
