```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Polygon
```

```python
y = np.array([ [1,1], [2,3], [3,1] ])
p = Polygon(y, facecolor='blue', edgecolor='magenta', alpha=0.3)

fig, ax = plt.subplots()
ax.set_xlim([0,4])
ax.set_ylim([0,4])
ax.add_patch(p)

plt.show()
```

### Exercise

#### $$ f(x) = -x^2  $$

```python
z = np.array([ [-0.5,-4.0], [-0.5,-2.5], [0.5,-2.5], [0.5,-4.0] ])
t = Polygon(z, facecolor='k', edgecolor='k', alpha=1)

x = np.linspace(-2,2,101)
f_x = -x**2

pnts = np.vstack((x, f_x)).T
pg = Polygon(pnts, facecolor='g', edgecolor='k', alpha=0.2)

fig, ax = plt.subplots()
ax.set_xlim([-2.2,2.2])
ax.set_ylim([-4.2,0.2])
ax.add_patch(t)
ax.add_patch(pg)
plt.axis('off')

plt.plot(x,f_x, color='k')
plt.plot([-2,2], [-4,-4], color='k')
plt.show()
```
