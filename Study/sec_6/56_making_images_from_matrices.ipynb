```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = [ [1,2], [1,4] ]

plt.imshow(A)

plt.xticks([0.2,0.8])
plt.yticks([0.85,1])

plt.show()
```

```python
A = np.zeros((10, 14))

for i in range(np.shape(A)[0]):
    for j in range(np.shape(A)[1]):
        A[i,j] = 3*i - 4*j
        
plt.imshow(A)
plt.plot([0,3],[8,2], 'r', linewidth=2)

for i in range(np.shape(A)[0]):
    for j in range(np.shape(A)[1]):
        plt.text(j, i, int(A[i,j]), 
                 horizontalalignment='center', 
                 verticalalignment='center')
        
plt.set_cmap('twilight_shifted')
plt.show()
```

## $$ c_{ij} = -1^{i+j}  $$

```python
N = 10
c = np.zeros((N, N))

for i in range(N):
    for j in range(N):
        c[i,j] = (-1)**(i+j)

plt.imshow(c)
plt.set_cmap('gray')
plt.axis('off')
plt.show()
```
