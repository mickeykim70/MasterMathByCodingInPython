### $$ y_1 = sin(\theta + cos(\theta))$$
### $$ y_2 = cos(\theta + sin(\theta))$$ 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
theta = np.linspace(-2*np.pi, 2*np.pi, 360)


y1 = np.sin(theta + np.cos(theta))
y2 = np.cos(theta + np.sin(theta))


plt.plot(theta, y1, label=f'$sin(\\theta + cos(\\theta))$')
plt.plot(theta, y2, label=f'$cos(\\theta + sin(\\theta))$')

plt.legend()

plt.show()
```
