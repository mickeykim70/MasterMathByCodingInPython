```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nums = np.random.rand(1000)

minval = 2
maxval = 17

nums = nums*(maxval-minval) + minval

plt.plot(nums, 'o')
plt.show()

plt.hist(nums)
plt.show()
```

#### Generate a Normal distribution with:
### mean = 15, standard deviation = 4.3

```python
desired_mean = 15
desired_std = 4.3

norm_num = np.random.randn(10000)*desired_std + desired_mean

plt.subplot(1,2,1)
plt.plot(norm_num, 's')

plt.subplot(122)
plt.hist(norm_num)

print(f'{np.mean(norm_num)}, {np.std(norm_num)}')
```
