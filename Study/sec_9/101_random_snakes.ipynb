Exercise: Random snakes from trigonometry

### $$ x = cos(r_1\theta)$$
### $$ y = sin(r_2\theta)$$

### $$ r_1, r_2  = random(0,1) $$ 
#### <center> using np.append </center>

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
theta = np.linspace(0, 8*np.pi, 1000)

r1 = np.random.rand(1)
r2 = np.random.rand(1)

x = np.cos(theta * r1)
y = np.sin(theta * r2)

# x = np.append(x, x[0])
# y = np.append(y, y[0])

plt.plot(x, y)

plt.axis('square')
plt.axis([-1.1, 1.1, -1.1, 1.1])
plt.axis('off')

plt.title(f'{{$r_1= {r1[0]:.2f}$}}, {{$r_2= {r2[0]:.2f}$}}')

plt.show()
```
