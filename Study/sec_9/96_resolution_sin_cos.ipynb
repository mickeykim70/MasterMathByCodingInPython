```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 6*np.pi, 400)

plt.plot(x, np.sin(x), 'r', label=f'$sin(\\theta)$')
plt.plot(x, np.cos(x), 'b', label=f'$cos(\\theta)$')

plt.legend()
plt.xlabel('radians ($\\theta$)')
plt.ylabel('function value')
plt.show()
```

```python
th = np.linspace(0, 10*np.pi, 400)

plt.plot(th, np.tan(th), 'k', label=f'$tan(\\theta)$')

plt.ylim([-100,100])

plt.legend()
plt.show()
```
