## $$ me^{ik}$$
#### <center> Random m's and k's </center>
<br>

#### <center> 40% pink, 40% purple and 20% green lines </center>

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_vectors = 200
m_s = np.random.rand(n_vectors)
k_s = np.random.rand(n_vectors) * 2*np.pi

for i in range(n_vectors):
    
    r = np.random.rand()
    
    if r < 0.4:
        clr = [1, .2, .7]   # pink
    elif r > 0.4 and r < 0.8:
        clr = [.7, .2, 1]   # purple
    else:
        clr = [0, 1, 0]   # green
    
    plt.polar( [0, k_s[i] ], [0, m_s[i] ], color=clr )

plt.axis('off')
plt.show()
```
