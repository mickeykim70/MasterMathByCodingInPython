## Euler's Formula

## $$ e^{ik} = cos(k)+isin(k)$$


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
k = np.pi/4  # angle
m = 2.3      # magnitude

eul = m * np.exp( 1j*k )  # Left hand side

# cos imaginary sin
cis = m * ( np.cos(k) + 1j*np.sin(k) )  # Right hand side


print(cis)
print(eul)
```

```python
mag = np.abs(eul)
ang = np.angle(eul)

print(mag)
print(ang)
```

```python
plt.polar( [0,ang], [0,mag], color='r' )
plt.polar( k, m, 'bo')


plt.show()
```

## $$ me^{i\theta}, \quad \; m = 5, \quad \theta=0.93$$
### cosine part: 3
### sine part: 4

```python
def eulerFromCosSin():
    
    real_num = eval( input('cosine part: '))
    imag_num = eval( input('sine part: '))
    
    m = np.sqrt( real_num**2 + imag_num**2 )
    k = np.arctan2(imag_num, real_num)
    
    plt.polar([0,k], [0,m], 'r-')
    
    plt.title(f'$ me^{{i\phi}}, {m=}, {{\phi}}={k:.2f} $')
    plt.thetagrids([0,45, 130, 155, 270])
    plt.show()
```

```python
eulerFromCosSin()
```
