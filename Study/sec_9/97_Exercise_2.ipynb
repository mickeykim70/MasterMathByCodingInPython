$$ sin(\theta) + cos(\theta), \quad at \quad \theta = \frac{7\pi}{6}$$ 

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy.printing.pretty.pretty_symbology import line_width
```

```python
# Graph unit circle
theta = np.linspace(0, 2*np.pi, 360)
theta_1 = (7/6)*np.pi

x = np.cos(theta)
y = np.sin(theta)


# Plot the unit circle
plt.plot([-1.1,1.1],[0,0],'--', color='gray')
plt.plot([0,0],[-1.1,1.1],'--', color='gray')
plt.plot(x, y, color='gray')


# Plot the arc
arc = np.linspace(0, theta_1, 300)

plt.plot(np.cos(arc), np.sin(arc), 'k', linewidth=2, label=f'$sin(\\theta_1)$')
plt.plot( [0,np.cos(theta_1)], [0,np.sin(theta_1)], 'k--') # plotting POINT
plt.plot( [0,np.cos(theta_1)], [0,0], color='green' )
plt.text(np.cos(theta_1)/2, 0 ,'a')
plt.plot( [np.cos(theta_1),np.cos(theta_1)], [0,np.sin(theta_1)], color='red' )

plt.axis('equal')
plt.legend()
plt.show()

```
