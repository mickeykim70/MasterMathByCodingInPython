```python
import numpy as np
import matplotlib.pyplot as plt

#  
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

#  
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# imshow 
im1 = ax1.imshow(Z, extent=[-5, 5, -5, 5], origin='lower', aspect='auto', cmap='viridis')
ax1.set_title('imshow')
plt.colorbar(im1, ax=ax1)

# pcolormesh 
im2 = ax2.pcolormesh(X, Y, Z, cmap='viridis')
ax2.set_title('pcolormesh')
plt.colorbar(im2, ax=ax2)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#  
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = x**2

# subplot 
plt.figure(figsize=(10, 8))

plt.subplot(221)  # 2 2 1
plt.plot(x, y1)
plt.title('Sine')

plt.subplot(222)  # 2 2 2
plt.plot(x, y2)
plt.title('Cosine')

plt.subplot(223)  # 2 2 3
plt.plot(x, y3)
plt.title('Tangent')

plt.subplot(224)  # 2 2 4
plt.plot(x, y4)
plt.title('Square')

plt.tight_layout()
plt.show()

# subplots 
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

axs[0, 0].plot(x, y1)
axs[0, 0].set_title('Sine')

axs[0, 1].plot(x, y2)
axs[0, 1].set_title('Cosine')

axs[1, 0].plot(x, y3)
axs[1, 0].set_title('Tangent')

axs[1, 1].plot(x, y4)
axs[1, 1].set_title('Square')

plt.tight_layout()
plt.show()
```

```python
x = np.linspace(-4,2,500)
y = np.linspace(-2,4,500)

X,Y = np.meshgrid(x,y)
s = 2

gauss2d = np.exp( -(X**2+Y**2)/s )

# plt.imshow(gauss2d)
plt.pcolormesh(gauss2d)
plt.axis('square')
plt.show()
```
