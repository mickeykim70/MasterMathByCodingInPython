#### Ellipses
### $$ \frac{(y-k)^2}{b^2} + \frac{(x-h)^2}{a^2} - 1 = 0$$
#### $$ Center: (h,k) \quad  Radii: a, b $$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
h = 1
k = 2
a = 2
b = 3

axlim = np.max( (a,b) ) + np.max( (abs(h),abs(k)) )
x = np.linspace(-axlim-1, axlim+1, 100)

X,Y = np.meshgrid(x,x)

# Create the function
ellipse = (Y-k)**2/b**2 + (X-h)**2/a**2 - 1


# Plot
plt.contour(X,Y,ellipse,0)
plt.plot(h,k,'go')
plt.axis('equal')

plt.show()
```

### Vary k from -4 to +4
### Vary a from |-4| to +4

#### h=0, b=4

```python
n = 26
h = 0
k = np.linspace(-4, 4, n)
a = abs(k)
b = 4

x = np.linspace(-8, 8, 100)
X,Y = np.meshgrid(x,x)

for i in range(n):
    F = (X-h)**2/a[i]**2 + (Y-k[i])**2/b**2 - 1
    plt.contour(X,Y,F,0)
plt.axis('equal') 
plt.show()
```
