#### Hyperbolas
### $$ \frac{(y-k)^2}{b^2} - \frac{(x-h)^2}{a^2} - 1 = 0$$
#### $$ Center: (h,k) \quad  Radii: a, b $$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 1
b = .5
h = 1
k = 2

axlim = 2*np.max( (a,b) ) + np.max( (abs(h),abs(k)) )
x = np.linspace(-axlim,axlim,100)
X,Y = np.meshgrid(x,x)

hyperbola = (X-h)**2/a**2 - (Y-k)**2/b**2 - 1

plt.contour(X,Y,hyperbola,0)
plt.plot(h,k,'go')


plt.plot( [-axlim,axlim], [0,0], 'k--', color=[0.8,0.8,0.8] )
plt.plot( [0,0], [-axlim,axlim], 'k--', color=[0.8,0.8,0.8] )

plt.axis('equal')
plt.show()
```
