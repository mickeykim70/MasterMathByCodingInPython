### Parabolas: graph and maths

### $$ 4p(y-k)=(x-h)^2  $$
### $$ y = \frac{1}{4p}(x-h)^2 + k $$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import label
```

```python
# Parameters
a = 1    # 1/4p
h = 1
k = -2
n = 101

# x-axis points to evaluate the function
x = np.linspace(h-5, h+5, n)

# Create the function
y = a*(x-h)**2 + k

plt.plot(x, y)
plt.grid()
plt.axis('square')
plt.show()
```

#### Exercise

### $$ Vertex: (h,k) $$
### $$ Focus: (h,k+p) $$
### $$ Directrix: y = k - p $$

```python
# Parameters
a = 1    # 1/4p... p = 4
p = 1/(4*a)
h = 1
k = -2
n = 101

# x-axis points to evaluate the function
x = np.linspace(h-2, h+2, n)

# Create the function
y = a*(x-h)**2 + k

# Plot the parabola
plt.plot(x, y, label='Parabola')

# plot the vertex
plt.plot(h, k, 'ro', label='Vertex')

# plot the focus
plt.plot(h, (k+p), 'go', label='Focus')

# plot the directrix
d = k-p
plt.plot( [x[0],x[-1]], [d,d], label='Directrix')

plt.grid()
# plt.axis('square')
plt.legend()
plt.show()
```
