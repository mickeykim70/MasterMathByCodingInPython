## Circles: graphs and maths

### $$ (y-k)^2 + (x-h)^2 = r^2 $$

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import pcolormesh
```

```python
h = 2
k = -3
r = 3

axlim = r + np.max((abs(h), abs(k)))
x = np.linspace(-axlim, axlim, 100)
X,Y = np.meshgrid(x,x)

Fxy = (X-h)**2 + (Y-k)**2 - (r**2)

plt.contour(X, Y, Fxy, 0)

plt.plot([-axlim,axlim],[0,0], 'k--')
plt.plot([0,0], [-axlim,axlim],'k--')
plt.plot(h,k,'go')

plt.gca().set_aspect('equal')
plt.show()
```

#### Exercise
Circle centers: -1.5 and + 1.5 <br>
Radii: .5 to 3

```python
# Parameters
h = [-1.5, 1.5]

x = np.linspace(-5, 5, 100)
X,Y = np.meshgrid(x,x)
radii = np.linspace(.5, 3, 10)

for radius in radii:
    for h_i in h:
        Fxy = (X-h_i)**2 + Y**2 - radius**2
        plt.contour(X,Y,Fxy,0,colors=[radius/5,radius/5,radius/5])
        

plt.gca().set_aspect('equal')

plt.plot(h, [0,0], 'k')

plt.show()


```
