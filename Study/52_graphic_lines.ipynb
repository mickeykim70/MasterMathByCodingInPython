```python
from cProfile import label

import numpy as np
import matplotlib.pyplot as plt
```

```python
# y = mx + b

x = [-5, 5]
m = 2
b = 1

y = [m * xi + b for xi in x]

plt.plot(x, y, label=f'y = {m}x + {b}')

plt.axis('equal')

axis = plt.gca()
print(axis.get_xlim(), axis.get_ylim())
plt.plot(axis.get_xlim(),[0,0], 'k--')
plt.plot([0,0], axis.get_ylim(), 'k--')
plt.xlim(x)
plt.ylim(x)
plt.grid()
plt.legend()
plt.show()
```

```python
# y = 0.7x -2
# y = -1.25x + 0.75

x = [-5, 5]
m = [0.7, -1.25]
b = [-2, 0.75]

for i in range(0, len(m)):
    y = m[i] * np.array(x) + b[i]
    # print(np.array(x[i]))
    # print(y)
    plt.plot(x, y, label=f'y = {m[i]}x + {b[i]}')
    
plt.axis('equal')

axis = plt.gca()
print(axis.get_xlim(), axis.get_ylim())
plt.plot(axis.get_xlim(),[0,0], 'k--')
plt.plot([0,0], axis.get_ylim(), 'k--')
plt.xlim(x)
plt.ylim(x)
plt.grid()
plt.legend()
plt.show()
    
```
