graph z1 + z2, z1z2

```python
z1 = -3 + 1j
z2 = -1 + 1j 

sum_z1z2 = z1 + z2
mul_z1z2 = z1 * z2

plt.plot([0,z1.real], [0,z1.imag], 'b-', label='z1')
plt.plot([0,z2.real], [0,z2.imag], 'r-', label='z2')
plt.plot([0,sum_z1z2.real], [0,sum_z1z2.imag], 'y-', label='z1+z2')
plt.plot([0,mul_z1z2.real], [0,mul_z1z2.imag], 'm-', label='z1*z2')


plt.axis([-5,5, -5,5])
plt.grid(True)
plt.legend()
plt.show()
```
