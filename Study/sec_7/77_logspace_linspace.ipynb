```python
a = 2
b = 100
n = 50

x = np.linspace(a, b, n)
log_x = np.logspace(np.log10(a), np.log10(b), n)
# log_x = np.logspace(a, b, n)

f_lin = x
f_log = log_x

plt.plot(x, f_lin, label='Linear')
plt.plot(x, f_log, label='Log-space')

plt.axis('square')
plt.legend()
plt.show()
```
