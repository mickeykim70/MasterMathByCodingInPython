### $$
f(x) = x^2
$$
f(x), f'(x) graph

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2, 2, 101)

fx = x**2
dx = x[1] - x[0]
df = np.diff(fx)/dx

plt.plot(x, fx, c='m', label=f'f(x)')
plt.plot(x[:-1], df, c='r', label=f'df/dx')


plt.axis([-2,2,-1,2])
plt.grid(True)
plt.legend()
plt.show()
```
