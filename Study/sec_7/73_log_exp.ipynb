```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-3, 3, 99)

plt.plot(x, np.exp(x))
plt.xlabel('x')
plt.ylabel('$e^x$')
plt.show()

plt.plot(x, np.log(x))
plt.xlabel('x')
plt.ylabel('$log(x)$')
plt.show()
```

### $$ y_1 = log(e^x)$$
### $$ y_2 = e^{log(x)}$$

```python
x = np.linspace(0.0001, 10, 10)

y1 = np.log( np.exp(x) )
y2 = np.exp( np.log(x) )

plt.plot(x, y1, label='$y = log(e^x)$')

plt.plot(x, y2, 'o',label='$y = e^{log(x)}$')

plt.legend()
plt.axis('square')
plt.show()
```
