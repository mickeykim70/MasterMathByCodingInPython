### Roots of unity

거듭제곱하여 1이 되는 복소수 <br> 
   - $x^n = 1$의 근이 되는 $x$를 1의 제곱근 <br>
   
### $$ z^n = 1, z \in \mathbb{C} $$
### $$ z = e^{2\pi ik/n}, \quad where \; k = 0, 1, 2, ... n-1 $$

```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
from IPython.display import display, Math
```

```python
n = 4

for k in range(0,n):
    root = np.exp( 2*np.pi*1j*k/n )
    print(root.real)
    print(root.imag)
```

```python
n = 4

for k in range(n):
    z = sym.exp(2*sym.pi*sym.I*k/n)
    display(Math(f"({sym.latex(z)})^{{{n}}} \\Rightarrow {sym.latex(z**n)}"))
```

```python
# Plot~~
n = 9

for k in range(n):
    root = np.exp( 2*np.pi*1j*k/n )
    plt.plot([0,root.real],[0,root.imag])

plt.axis('square')
plt.show()
```
