```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2*np.pi, 200)

plt.plot(np.cos(x), np.sin(x))

plt.plot([-1.2,1.2], [0,0], ls='--', c=[0.8,0.8,0.8])
plt.plot([0,0], [-1.2,1.2], ls='--', c=[0.8,0.8,0.8])

angle = np.pi / 4
plt.plot([0, np.cos(angle)], [0, np.sin(angle)])


plt.axis('square')
plt.xlabel('cos(x)')
plt.ylabel('sin(x)')
plt.show()
```
